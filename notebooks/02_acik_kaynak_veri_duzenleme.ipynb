```python
mahalle_nufus = pd.read_excel(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\raw\bursa_mahalle_nufus.xls")
mahalle_nufus.head(10)
```

```python
mahalle_nufus = mahalle_nufus.dropna(subset=["Unnamed: 1", "Sütunlar"])
mahalle_nufus.head(10)
```

```python
mahalle_nufus = mahalle_nufus[~mahalle_nufus["Unnamed: 1"].str.contains("Belediye|Satırlar|Ölçüm", na=False)]
```

```python
mahalle_nufus = mahalle_nufus.rename(columns={"Unnamed: 1": "adres", "Sütunlar": "nufus", "Unnamed: 0": "yil"})
```

```python
mahalle_nufus["mahalle"] = mahalle_nufus["adres"].str.extract(r"/([^/]+ Mah\.)")[0].str.replace(" Mah.", "", regex=False)
mahalle_nufus.head()
```

```python

mahalle_nufus["ilce"] = mahalle_nufus["adres"].str.extract(r"Bursa\(([^/]+)")
mahalle_nufus["il"] = "Bursa"
mahalle_nufus["nufus"] = mahalle_nufus["nufus"].astype(int)
mahalle_nufus['yil'] = 2024
mahalle_nufus["yil"] = mahalle_nufus["yil"].astype(int)
mahalle_nufus.head()
```

```python
mahalle_nufus["mahalle_kodu"] = mahalle_nufus["adres"].str.extract(r"[-](\d+)$")
mahalle_nufus.head()
```

```python
mahalle_nufus["mahalle"] = mahalle_nufus["mahalle"].str.strip()
mahalle_nufus["mahalle_kodu"] = mahalle_nufus["mahalle_kodu"].str.strip().astype(int)
mahalle_nufus.head()
```

```python
mahalle_nufus = mahalle_nufus[['yil', 'adres', 'il', 'ilce', 'mahalle', 'mahalle_kodu', 'nufus']]
mahalle_nufus.head()
```

```python
mahalle_nufus.to_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\mahalle_nufus_duzenli.csv", index=False, encoding="utf-8-sig")
```

```python
ilce_yas = pd.read_excel(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\raw\bursa_ilce_yas_dagilimi.xls")
ilce_yas.head(10)
```

```python
ilce_yas.columns = ilce_yas.iloc[1]

ilce_yas.columns = ilce_yas.columns.str.replace("-", "_")

ilce_yas.head()
```

```python
nan_indices = [i for i, col in enumerate(ilce_yas.columns) if pd.isna(col)]


if len(nan_indices) >= 2:
    ilce_yas.columns.values[nan_indices[0]] = "yil"
    ilce_yas.columns.values[nan_indices[1]] = "adres"
```

```python
ilce_yas.head()
```

```python
ilce_yas = ilce_yas.iloc[4:].reset_index(drop=True)
ilce_yas.head()
```

```python
ilce_yas["ilce"] = ilce_yas["adres"].str.extract(r"\((.*?)\)")
ilce_yas["ilce_kodu"] = ilce_yas["adres"].str.extract(r"-(\d+)")
ilce_yas["ilce"] = ilce_yas["ilce"].str.strip()
ilce_yas["ilce_kodu"] = ilce_yas["ilce_kodu"].str.strip().astype(int)
ilce_yas['yil'] = 2024


ilce_yas.head()
```

```python
ilce_yas["yas_0_14"] = ilce_yas[["0_4", "5_9", "10_14"]].sum(axis=1)
ilce_yas["yas_15_24"] = ilce_yas[["15_19", "20_24"]].sum(axis=1)
ilce_yas["yas_25_34"] = ilce_yas[["25_29", "30_34"]].sum(axis=1)
ilce_yas["yas_35_49"] = ilce_yas[["35_39", "40_44", "45_49"]].sum(axis=1)
ilce_yas["yas_50_64"] = ilce_yas[["50_54", "55_59", "60_64"]].sum(axis=1)
ilce_yas["yas_65_plus"] = ilce_yas[["65_69", "70_74", "75_79", "80_84", "85_89", "90+"]].sum(axis=1)

yas_sutunlari = ["0_4", "5_9", "10_14", "15_19", "20_24", "25_29", "30_34", "35_39", "40_44",
                 "45_49", "50_54", "55_59", "60_64", "65_69", "70_74", "75_79", "80_84", "85_89", "90+"]
ilce_yas = ilce_yas.drop(columns=yas_sutunlari)

ilce_yas.head()
```

```python
ilce_yas["il"] = "Bursa"
ilce_yas["tum_yas_toplam"] = ilce_yas[
    ["yas_0_14", "yas_15_24", "yas_25_34", "yas_35_49", "yas_50_64", "yas_65_plus"]
    ].sum(axis=1)
ilce_yas = ilce_yas[['yil', 'adres', 'il', 'ilce', 'ilce_kodu', "yas_0_14","yas_15_24","yas_25_34","yas_35_49","yas_50_64","yas_65_plus","tum_yas_toplam"]]
ilce_yas.head()
```

```python
ilce_yas.to_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\ilce_yas_duzenli.csv", index=False, encoding="utf-8-sig")
```
