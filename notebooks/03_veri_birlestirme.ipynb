```python
import geopandas as gpd
import pandas as pd

mahalle_gdf = gpd.read_file(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\mahalle_sinir_duzenli.geojson")
mahalle_nufus = pd.read_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\mahalle_nufus_duzenli.csv")

mahalle_joined = mahalle_gdf.merge(mahalle_nufus, on="mahalle_kodu", how="left")

mahalle_joined.head()
```

```python
mahalle_joined[mahalle_joined.isna().any(axis=1)]
```

```python
mahalle_joined["mahalle_y"] = mahalle_joined["mahalle_y"].fillna(mahalle_joined["mahalle_x"])

mahalle_joined.head()
```

```python
mahalle_joined["nufus_yogunlugu"] = mahalle_joined["nufus"] / (mahalle_joined["alan"] / 1_000_000)
```

```python
mahalle_joined = mahalle_joined.rename(columns={"il_x": "il","mahalle_y":"mahalle"})

kolonlar = [
    "yil",
    "il",
    "ilce",
    "mahalle",
    "mahalle_kodu",
    "uzunluk",
    "alan",
    "nufus",
    "nufus_yogunlugu",
    "geometry"
]

mahalle_joined = mahalle_joined[kolonlar]

mahalle_joined.head()
```

```python
mahalle_joined = mahalle_joined.sort_values(by="mahalle", ascending=True).reset_index(drop=True)
mahalle_joined.head()
```

```python
mahalle_joined["nufus"] = mahalle_joined["nufus"].astype("Int64")
mahalle_joined.head()
```

```python
mahalle_joined.to_file(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\mahalle_joined.geojson", driver="GeoJSON")
```

```python
import os
from dotenv import load_dotenv

load_dotenv()
db_url = os.getenv("DB_URL")

from sqlalchemy import create_engine 
engine = create_engine(db_url)
mahalle_joined.to_postgis("mahalle_joined", engine, if_exists="replace", index=False)
```

```python

ilce_gdf = gpd.read_file(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\ilce_sinir_duzenli.geojson")
ilce_nufus = pd.read_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\ilce_nufus_duzenli.csv")

ilce_gdf.head()
```

```python
ilce_gdf = ilce_gdf.sort_values(by="ilce").reset_index(drop=True)
ilce_gdf.head(50)
```

```python
ilce_nufus = ilce_nufus.sort_values(by="ilce").reset_index(drop=True)
ilce_nufus.head(50)
```

```python
ilceler_dogru = [
    "Büyükorhan",
    "Gemlik",
    "Gürsu",
    "Harmancık",
    "Karacabey",
    "Keles",
    "Kestel",
    "Mudanya",
    "Mustafakemalpaşa",
    "Nilüfer",
    "Orhaneli",
    "Orhangazi",
    "Osmangazi",
    "Yenişehir",
    "Yıldırım",
    "İnegöl",
    "İznik"
]

ilce_gdf["ilce"] = ilceler_dogru
```

```python
ilce_gdf.head(50)
```

```python
ilce_joined = ilce_gdf.merge(
    ilce_nufus,
    on=["il", "ilce"], 
    how="left"
)

ilce_joined.head()
```

```python
ilce_joined = ilce_joined.drop(columns=["ilce_kodu_x", "ilce_kodu_y"])
ilce_joined.head()
```

```python
ilce_yas = pd.read_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\ilce_yas_duzenli.csv")

ilce_yas.head()
```

```python
ilce_joined2 = ilce_joined.merge(
    ilce_yas,
    on=["yil", "il", "ilce"], 
    how="left"
)

ilce_joined2.head()
```

```python
ilce_joined2 = ilce_joined2.drop(columns=["adres_x", "adres_y"])

ilce_joined2["nufus_yogunlugu"] = ilce_joined2["nufus"] / (ilce_joined2["alan_5254"] / 1000000)
ilce_joined2.head()
```

```python
ilce_ses = pd.read_csv(r"C:\Users\Monster\Desktop\CBS_CALISMALARI\getir_case_study\data\processed\ilce_ses_duzenli.csv")

ilce_ses.head()
```

```python
ilce_final = ilce_joined2.merge(
    ilce_ses,
    on=["il", "ilce"], 
    how="left"
)
ilce_final.head()
```
